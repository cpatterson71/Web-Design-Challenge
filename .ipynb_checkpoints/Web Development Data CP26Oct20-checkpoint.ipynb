```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file = 'Resources/Medical_Conditions_update.csv'
```

```python
med_cond = pd.read_csv(file)
med_cond
```

```python
med_cond_df = pd.DataFrame(med_cond)
print(med_cond_df)
```

```python
x_value = med_cond_df['Medical_Condition']
y_value = med_cond_df['Children_Count']
z_value = med_cond_df['Adults_Count']
 
tick_locations = []
for x in x_value:
    tick_locations.append(x)
    
plt.xlim(-0.75, len(x_value)-.5)
plt.ylim(0, max(y_value) + 5)

plt.bar(x_value, y_value, facecolor="red", alpha=0.75, align="center")
plt.xticks(tick_locations, x_value)
plt.show()
 
```

```python
med_cond_df.to_csv('Resources/med_cond_df.csv')
```

```python
plt.title("% of Children Deaths based on Medical Condtion")
plt.pie(y_value, labels=x_value,
        autopct="%1.1f%%", shadow=True, startangle=90)
plt.axis("equal")
plt.show()
```

```python
plt.title("% of Adult Deaths based on Medical Condtion")
plt.pie(y_value, labels=x_value,
        autopct="%1.1f%%", shadow=True, startangle=90)
plt.axis("equal")
plt.show()
```
